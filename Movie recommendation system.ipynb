```python
#CREATING A MOVIE RECOMMENDATION SYSTEM
```

```python
df.head()
```

```python
movie_matrix = df.pivot_table(index = "user_id", columns = "title", values = "rating") 
```

```python
movie_matrix
```

```python
ratings.sort_values("count", ascending = False)
```

```python
star_wars_rating = movie_matrix["Star Wars (1977)"]
```

```python
star_wars_rating
```

```python
similar_to_star_wars = movie_matrix.corrwith(star_wars_rating)
```

```python
corr_star_wars = pd.DataFrame(similar_to_star_wars, columns = ["correlation"])
```

```python
corr_star_wars
```

```python
corr_star_wars.dropna(inplace = True)
```

```python
corr_star_wars.sort_values("correlation",ascending = False)
```

```python
corr_star_wars.sort_values("correlation",ascending = False).head(15)
```

```python
#adding a threshold
```

```python
 corr_star_wars = corr_star_wars.join(ratings["count"])
corr_star_wars.head()
```

```python
corr_star_wars[corr_star_wars["count"]>100].sort_values("correlation", ascending = False)
```

```python
#PREDICT FUNCTION
```

```python
def predict_movies(movie_name):
    movie_user_ratings = movie_matrix[movie_name]
    similar = movie_matrix.corrwith(movie_user_ratings)
    
    corr_movie = pd.DataFrame(similar, columns = ["correlation"])
    corr_movie.dropna(inplace = True)

    corr_movie = corr_movie.join(ratings["count"])
    prediction = corr_movie[corr_movie["count"]>100].sort_values("correlation",ascending = False)

    return prediction
    
```

```python
prediction = predict_movies("Crash (1996)")
```
